## Data and Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
data = pd.read_csv('data/penguins.csv')
data.head()
```

## Seaborn Styles

```python
sb.set_style( 'dark' )
sb.set_context( 'notebook' )
sb.set_palette( 
    'mako_r',
    #'Oranges_r'
)
#sb.palplot( sb.color_palette( 'Oranges_r', 11 ) )
sb.palplot( sb.color_palette( 'mako_r', 11 ) )
```

## Plots

### Distribution Plots

```python
sb.distplot( data['bill_depth_mm'] )
```

```python
sb.rugplot( data['bill_depth_mm'] )
```

```python
sb.kdeplot( data['bill_depth_mm'] )
```

```python
sb.FacetGrid(
    data,
    hue = "species",
    height=5
).map(
    sb.kdeplot,
    "bill_depth_mm"
).add_legend()
plt.ioff()
```

```python
sb.jointplot(
    x = "bill_length_mm",
    y = "bill_depth_mm", 
    data = data
)
```

```python
sb.pairplot( data )
```

```python
sb.scatterplot(
    x = "bill_length_mm",
    y = "bill_depth_mm", 
    data = data
)
```

## Statistical Models

```python
data.describe().plot(
    kind = "area",
    fontsize = 20, 
    figsize = ( 20, 8 ),
    table = False
)
plt.xlabel('Statistics')
plt.ylabel('Value')
plt.title("Statistics of PalmerPengins Dataset")
```

```python
sb.residplot(
    x = "bill_length_mm",
    y = "bill_depth_mm", 
    data = data
)
```

```python
sb.lmplot(
    x = "bill_length_mm",
    y = "bill_depth_mm", 
    data = data
)
```

```python
sb.heatmap(
    data.corr(),
    vmin = -1,
    vmax = 1,
    annot = True
)
```
